```python
# Cargar tablas de compras y juntarlas.

df_compras = list()
for a in range(1, 6):
    df_compras.append(pd.read_csv(f"Tablas/tabla_compras_{a}.csv", sep = ';'))

df_compras = pd.concat(df_compras)
df_compras
```
